## Grouping and Aggregation

```python
#average sold price for each age category
df.groupby('AGE')['SOLD PRICE'].mean()
```

```python
#sort by descending average sold price for each country category
df.groupby('COUNTRY')['SOLD PRICE'].mean().sort_values(ascending=False)
```

```python
#resetting indexes by average sold price
soldprice_by_age=df.groupby('AGE')['SOLD PRICE'].mean().reset_index()
soldprice_by_age
```

```python
#grouping by age, playing role and then getting avg SP of each
#getting avg SP for different playing roles in each age category
#resetting indexes by average sold price (SP is used and not any other columns as SP has unique values)
#used when the index is meaningless and needs to be reset to the default 
soldprice_by_agerole=df.groupby(['AGE','PLAYING ROLE'])['SOLD PRICE'].mean().reset_index()
soldprice_by_agerole
```

## Joining Dataframes

```python
#compare average auction price for different ages and playing role
#join DFs based on column Age
#taking outer join
soldprice_comparison = soldprice_by_agerole.merge(soldprice_by_age, on = 'AGE', how = 'outer')
soldprice_comparison
```

## Renaming Columns

```python
soldprice_comparison.rename(columns = {'SOLD PRICE_x':'SOLD PRICE_AGE_ROLE', 'SOLD PRICE_y':'SOLD PRICE_AGE'}, inplace = True)
```
